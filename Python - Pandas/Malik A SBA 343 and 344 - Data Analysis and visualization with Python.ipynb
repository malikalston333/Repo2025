```python
total_profits =  df["total_profit"]
total_profits.plot(kind='bar', color=['royalblue'])
plt.xlabel('Month number')
plt.ylabel('Total Profit')
plt.title('Company profit per month')
plt.show()
```

```python
total_profits =  df["total_profit"]
total_profits.plot(linestyle='dotted', linewidth='3', marker= 'o', markerfacecolor='black', color=['red'])
plt.xlabel('Month number')
plt.ylabel('Total Profit')
plt.title('Company profit per month')
plt.legend(loc='lower right', title='Profit data of last year')
plt.show()
```

```python
facecream_sales = df["facecream"]
facecream_sales.plot(marker= 'o')
facewash_sales = df["facewash"]
facewash_sales.plot(marker= 'o')
toothpaste_sales = df["toothpaste"]
toothpaste_sales.plot(marker= 'o')
bathingsoap_sales = df["bathingsoap"]
bathingsoap_sales.plot(marker= 'o')
shampoo_sales = df["shampoo"]
shampoo_sales.plot(marker= 'o')
moisturizer_sales = df["moisturizer"]
moisturizer_sales.plot(marker= 'o')
plt.xlabel('Month number')
plt.ylabel('Sales units in numbers')
plt.title('Sales Data')
plt.xticks(np.arange(0,13))
plt.legend()
plt.show()
```

```python
bathingsoap_sales = df["bathingsoap"]
month = df["month_number"]
plt.scatter(month, bathingsoap_sales, label='bathing soap Sales data')
plt.xlabel('Month number')
plt.ylabel('Number of units sold')
plt.title('Sales Data')
plt.grid(linestyle="--")
plt.xticks(np.arange(0,13))
plt.legend()
plt.show()
```
