# 7. Survival Count Bar Plot:
#    - Read the count of survivors and non-survivors.
#    - Show it using a bar plot.

```python
survival_counts = df['Survived'].value_counts() # Count of survivors
survival_counts.plot(kind='bar', color=['blue','purple'])
plt.xlabel('Survival Status (0 = No, 1 = Yes)')
plt.ylabel('Number of Passengers')
plt.title('Survival Count')
plt.show()
# Using Seaborn
# Create the count plot for the 'Survived' column
#sns.countplot(x='Survived', data=df, palette=['red', 'green'])

# Add labels and title
#plt.xlabel('Survival Status (0 = No, 1 = Yes)')
#plt.ylabel('Number of Passengers')
#plt.title('Survival Count')

# Show the plot
#plt.show()
```

# 8. Age Distribution Histogram:
#    - Read the age data.
#    - Show the distribution of ages using a histogram.

```python
plt.hist(df['Age'].dropna(), bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Number of Passengers')
plt.title('Age Distribution of Passengers')
#It excludes any missing values (dropna()).
#It divides the ages into 20 bins (bins=20).
# Show the plot
plt.show()
```

# 9. Survival Rate by Passenger Class (Pclass):
#    - Group the data by "Pclass" and calculate the survival rate for each class.
#    - Show it using a bar plot.

```python
# Group the data by "Pclass" and calculate the survival rate for each class
survival_rate_by_class = df.groupby('Pclass')['Survived'].mean()

# Display the survival rate by class
print(survival_rate_by_class)
```

```python
survival_rate_by_class.plot(kind='bar', color=['green', 'orange', 'red'])

# Add labels and title
plt.xlabel('Passenger Class (Pclass)')
plt.ylabel('Survival Rate')
plt.title('Survival Rate by Passenger Class')
# Show the plot
plt.show()

#survival_rate_by_class = df.groupby('Pclass')['Survived'].mean().reset_index()

# Plot the survival rate by class using Seaborn
#plt.figure(figsize=(8, 6))
#sns.barplot(x='Pclass', y='Survived', data=survival_rate_by_class, palette='muted')
# Add labels and title
#plt.xlabel('Passenger Class (Pclass)')
#plt.ylabel('Survival Rate')
#plt.title('Survival Rate by Passenger Class')

# Show the plot
#plt.show()
```

# 10. Age vs. Fare Scatter Plot:
#     - Create a scatter plot showing the relationship between "Age" and "Fare".


```python
df_clean = df[['Age','Fare']].dropna()
plt.scatter(df_clean['Age'],df_clean['Fare'],color = 'blue',alpha=0.6)
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Age vs. Fare Scatter Plot')
plt.show()
# alpha=1.0: Fully opaque (default  not see-through).
# alpha=0.0: Fully transparent (invisible).
# alpha=0.3 to 0.7: Semi-transparent  useful when plotting many overlapping points.
```

# 11. Survival Count by Sex:
#     - Read the count of survivors and non-survivors grouped by "Sex".
#     - Show it using a count plot from the Seaborn library.


```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Sex', hue='Survived', data=df, palette=['red', 'green'])

# Add labels and title
plt.xlabel('Sex')
plt.ylabel('Number of Passengers')
plt.title('Survival Count by Sex')
plt.legend(title='Survived', labels=['No (0)', 'Yes (1)'])

# Show the plot
plt.show()
```
